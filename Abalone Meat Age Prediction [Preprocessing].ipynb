```python
import numpy as np
abalone["Rings"]=abalone["Rings"].astype(np.float64)
```

```python
from sklearn.preprocessing import LabelEncoder
LB = LabelEncoder()
abalone['Sex'] = LB.fit_transform(abalone['Sex'])
```

```python
Y = abalone['Age']
abalone_ready = abalone.drop('Age', axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(abalone_ready, Y, test_size=0.2, random_state=20)
```
